### Q1: For the given dataset, calculate the variance, standard deviation, and interquartile range (IQR) of the salaries of employees in the HR and Marketing departments.


```python
variance_hr_sales_dept = df[df['Department'].isin(['Human Resources', 'Sales'])]['MonthlyIncome'].var()
print("Variance of Monthly Income for HR and Sales in Department:", variance_hr_sales_dept)
```

```python
variance_hr_sales_ef = df[df['EducationField'].isin(['Human Resources', 'Marketing'])]['MonthlyIncome'].var()
print("Variance of Monthly Income for HR and Marketing in EducatonalField:", variance_hr_sales_ef)
```

```python
standard_deviation = np.sqrt(variance_hr_sales_dept)
print("Standard Deviation of Monthly Income for HR and Sales in Department:", standard_deviation)
```

```python
standard_deviation = np.sqrt(variance_hr_sales_ef)
print("Standard Deviation of Monthly Income for HR and Sales in Department:", standard_deviation)
```

```python
hr_sales_dept = df[df['Department'].isin(['Human Resources', 'Sales'])]
Q1 = hr_sales_dept['MonthlyIncome'].quantile(0.25)
Q3 = hr_sales_dept['MonthlyIncome'].quantile(0.75)
IQR = Q3 - Q1
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```

```python
hr_sales_ef = df[df['EducationField'].isin(['Human Resources', 'Marketing'])]
Q1 = hr_sales_ef['MonthlyIncome'].quantile(0.25)
Q3 = hr_sales_ef['MonthlyIncome'].quantile(0.75)
IQR = Q3 - Q1
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```

### Q2: What is the range of experience in the IT department?
 - There are three only three departments in the dataset
 - Research & Development,Sales,Human Resources
 - Calculating range of experience

```python
grouped_experience = df.groupby('Department')['TotalWorkingYears']
max_experience = grouped_experience.max()
min_experience = grouped_experience.min()
experience_range = max_experience - min_experience
experience_stats = pd.DataFrame({
    'Max Experience': max_experience,
    'Min Experience': min_experience,
    'Range of Experience': experience_range
})
print(experience_stats)
```

### Q3: Create a boxplot to display the experience of each department.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Department', y='TotalWorkingYears', data=df)

plt.title('Experience (Total Working Years) by Department')
plt.xlabel('Department')
plt.ylabel('Total Working Years')

plt.xticks(rotation=45)
plt.show()
```

### Q4: Which feature has the highest standard deviation in the dataset?


```python
numeric_df = df.select_dtypes(include=np.number)
std_devs = numeric_df.std()

max_std_feature = std_devs.idxmax()
max_std_value = std_devs.max()

print(f"Feature with the highest standard deviation: {max_std_feature}")
print(f"Standard Deviation: {max_std_value}")
```
